#### Ejercicio 10. ¿Son los promedios de precio por metro cuadrado (price/m2) de "Valdemorillo" y "Galapagar" los mismos? (★★☆)

Imprime ambos promedios de precio y escribe una conclusion acerca de ellos.

Pista: Crea una nueva columna llamada `pps` (price per square) y luego analiza los valore.s

```python
import pandas as pd

# Cargar el dataset
data = pd.read_csv('assets/real_estate.csv', sep=';')

valdemorillo_data = data[data['level5'] == 'Valdemorillo']
galapagar_data = data[data['level5'] == 'Galapagar']

valdemorillo_data['pps'] = valdemorillo_data['price'] / valdemorillo_data['surface']
galapagar_data['pps'] = galapagar_data['price'] / galapagar_data['surface']

media_pps_valdemorillo = valdemorillo_data['pps'].mean()
media_pps_galapagar = galapagar_data['pps'].mean()

print("Precio promedio por metro cuadrado en Valdemorillo:", media_pps_valdemorillo)
print("Precio promedio por metro cuadrado en Galapagar:", media_pps_galapagar)
```

#### Exercise 18. Normalize the variable of prices for each population and plot the 4 histograms in the same plot (you must use the subset obtained in the question 14) (★★★)

For the normalization method you can use the one you consider to, there is not a single correct answer to this question. Print the plot and write in the Markdown cell a brief analysis about the plot

Hint: You can help yourself reviewing the multihist demo of Matplotlib

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('assets/real_estate.csv', sep=';')

cinturon_sur_poblaciones = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]

cinturon_sur_data = data[data['level5'].isin(cinturon_sur_poblaciones)]

mean_prices = cinturon_sur_data.groupby('level5')['price'].mean()
std_prices = cinturon_sur_data.groupby('level5')['price'].std()

normalized_prices = (cinturon_sur_data['price'] - cinturon_sur_data['level5'].map(mean_prices)) / cinturon_sur_data['level5'].map(std_prices)

plt.figure(figsize=(10, 6))
for population in cinturon_sur_poblaciones:
    plt.hist(normalized_prices[cinturon_sur_data['level5'] == population], bins=20, alpha=0.5, label=population)

plt.title('Normalized Prices Histogram for Populations in Cinturón Sur of Madrid')
plt.xlabel('Normalized Price')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()
```

#### Exercise 20. Make the same plot for 4 different populations (level5 column) and rearrange them on the same graph? You must use the subset obtained in the question 14 (★★☆) 
Hint: make a scatter plot of each population using subplots.

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('assets/real_estate.csv', sep=';')

cinturon_sur_poblaciones = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]

cinturon_sur_data = data[data['level5'].isin(cinturon_sur_poblaciones)]

fig, axes = plt.subplots(2, 2, figsize=(12, 8))

for i, population in enumerate(cinturon_sur_poblaciones):
    ax = axes[i // 2, i % 2]
    population_data = cinturon_sur_data[cinturon_sur_data['level5'] == population]
    ax.scatter(population_data['surface'], population_data['price'], label=population, alpha=0.7)
    ax.set_title(population)
    ax.set_xlabel('Surface')
    ax.set_ylabel('Price')

plt.tight_layout()

plt.show()
```

```python
from ipyleaflet import Map, Polyline, basemaps

madrid_map = Map(center=(40.4168, -3.7038), zoom=10, min_zoom=1, max_zoom=20, basemap=basemaps.OpenStreetMap.Mapnik)

coordenadas = [
    {"nombre": "Fuenlabrada", "latitud": 40.28382, "longitud": -3.79462},
    {"nombre": "Getafe", "latitud": 40.30817, "longitud": -3.73479},
    {"nombre": "Leganés", "latitud": 40.32718, "longitud": -3.7635},
    {"nombre": "Móstoles", "latitud": 40.32233, "longitud": -3.86496},
    {"nombre": "Alcorcón", "latitud": 40.34968, "longitud": -3.82835}
]

colores = {
    "Fuenlabrada": "red",
    "Getafe": "blue",
    "Leganés": "green",
    "Móstoles": "purple",
    "Alcorcón": "orange"
}
for i in range(len(coordenadas)-1):
    punto_actual = coordenadas[i]
    siguiente_punto = coordenadas[i+1]
    linea = Polyline(
        locations=[(punto_actual["latitud"], punto_actual["longitud"]),
                   (siguiente_punto["latitud"], siguiente_punto["longitud"])],
        color=colores[punto_actual["nombre"]]
    )
    madrid_map.add_layer(linea)

madrid_map
```
